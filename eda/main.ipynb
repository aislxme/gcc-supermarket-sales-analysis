```python
numerical_cols = ['Unit price', 'Quantity', 'Total','Rating']

for col in numerical_cols:
    plt.title(f'Before Cleaning: {col}')
    plt.boxplot(df[col])
    plt.show()

    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR

    df=df[(df[col]>=lower_limit) & (df[col]<=upper_limit)]

    plt.title(f'After Cleaning:{col}')
    plt.boxplot(df[col])
    plt.show()
```

EDA

```python
# Revenue overview
print("📊 KEY METRICS:")
print(f"Total Revenue: ${df['Total'].sum():,.0f}")
print(f"Avg Order: ${df['Total'].mean():.0f}")
print(f"Total Orders: {len(df)}")

```

```python
# Top performers
print("\n🏆 BY BRANCH:")
print(df.groupby('Branch')['Total'].agg(['count', 'sum', 'mean']).round(0))

print("\n🏆 BY CITY:")
print(df.groupby('City')['Total'].agg(['count', 'sum', 'mean']).round(0))
```

```python
# Revenue share by city
city_sales = df.groupby('City')['Total'].sum().sort_values(ascending=False)
city_share = ((city_sales / city_sales.sum()) * 100).round(2)

print("Revenue Share by City:")
for city, pct in city_share.items():
    print(f"{city}: {pct}% revenue share")

# Top product category
product_sales = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)

print("\nRevenue by Product Line:")
print(product_sales)

top_product = product_sales.idxmax()
print(f"\nTop category = {top_product}")
```

Charts

```python
# 1. Revenue by City (GCC-style)
plt.figure(figsize=(10,6))
sns.barplot(data=df, x='City', y='Total')
plt.title('Revenue by GCC City (Jan-Mar 2019)', fontsize=14, fontweight='bold')
plt.ylabel('Total Revenue ($)')
plt.show()
```

```python
# 2. Top 5 Products
top_products = df.groupby('Product line')['Total'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(10,6))
top_products.plot(kind='bar')
plt.title('Top 5 Product Categories', fontweight='bold')
plt.xticks(rotation=45)
plt.ylabel('Revenue ($)')
plt.show()
```

```python
# 3. Hourly Heatmap (Business Gold!)
df['Hour'] = df['Time'].str[:2].astype(int)
hourly = df.pivot_table(values='Total', index='Hour', columns='Branch', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(hourly, annot=True, fmt='.0f', cmap='YlOrRd')
plt.title('Hourly Sales Heatmap by Branch')
plt.show()
```

```python
import pandas as pd

# 1. Revenue by City (Real percentages)
city_revenue = df.groupby('City')['Total'].sum().sort_values(ascending=False)
city_pct = (city_revenue / city_revenue.sum() * 100).round(1)

print("🏙️ CITY REVENUE SHARE:")
for city, pct in city_pct.items():
    print(f"  {city}: {pct}%")
print(f"  Total: ${city_revenue.sum():,.0f}\n")

# 2. Peak sales hour (Real peak time)
df['Hour'] = df['Time'].str[:2].astype(int)
hourly_sales = df.groupby('Hour')['Total'].sum()
peak_hour = hourly_sales.idxmax()

print(f"⏰ PEAK SALES: Hour {peak_hour}:00 (${hourly_sales[peak_hour]:,.0f})")

# 3. Top product category (Real #1)
product_revenue = df.groupby('Product line')['Total'].sum().sort_values(ascending=False)
top_category = product_revenue.index[0]
top_category_pct = (product_revenue.iloc[0] / product_revenue.sum() * 100).round(1)

print(f"\n🛍️ TOP CATEGORY: {top_category} ({top_category_pct}% of total)")

# 4. Payment method distribution (Real %)
payment_dist = df['Payment'].value_counts(normalize=True) * 100
top_payment = payment_dist.index[0]
top_payment_pct = payment_dist.iloc[0].round(1)

print(f"\n💳 TOP PAYMENT: {top_payment} ({top_payment_pct}%)")

# 5. Member vs Normal spending (Real difference)
customer_spend = df.groupby('Customer type')['Total'].agg(['mean', 'count'])
member_avg = customer_spend.loc['Member', 'mean']
normal_avg = customer_spend.loc['Normal', 'mean']
spend_diff_pct = ((member_avg / normal_avg - 1) * 100).round(1)

print(f"\n👥 CUSTOMER SPEND:")
print(f"  Member avg: ${member_avg:.0f}")
print(f"  Normal avg: ${normal_avg:.0f}")
print(f"  Members spend {spend_diff_pct}% more")
```
