```python
numerical_cols = ['Unit price', 'Quantity', 'Total','Rating']

for col in numerical_cols:
    plt.title(f'Before Cleaning: {col}')
    plt.boxplot(df[col])
    plt.show()

    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    lower_limit=Q1-1.5*IQR
    upper_limit=Q3+1.5*IQR

    df=df[(df[col]>=lower_limit) & (df[col]<=upper_limit)]

    plt.title(f'After Cleaning:{col}')
    plt.boxplot(df[col])
    plt.show()
```

```python
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Month'] = df['Date'].dt.month
df['DayOfWeek'] = df['Date'].dt.dayofweek
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.hour  # FIXED for hh:mm:ss
df['Peak_Hour'] = df['Hour'].apply(lambda x: 1 if x in [10,11,12,13,14,18,19,20] else 0)
```

```python
df.head()
```

```python
# Financial
df['Avg_Price_Per_Unit'] = df['Total'] / df['Quantity']
df['Profit_Margin'] = (df['Total'] - df['cogs']) / df['Total'] * 100
df['High_Value_Customer'] = (df['Total'] > df['Total'].quantile(0.8)).astype(int)
```

```python
df
```

```python
numerical_cols = ['Unit price', 'Quantity', 'Total', 'Rating']
plt.figure(figsize=(12, 8))

for i, col in enumerate(numerical_cols, 1):
    plt.subplot(2, 3, i)
    plt.hist(df[col], bins=20, color='pink', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
